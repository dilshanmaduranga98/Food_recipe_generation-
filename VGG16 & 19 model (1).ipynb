# **-- VGG 16 Model --**

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import pandas as pd
from pathlib import Path
import os.path
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_dir = Path('/content/drive/My Drive/Final Project/Fast food Dataset/Train')
train_filepath = list(train_dir.glob(r'**/*.jpeg'))
```

```python
test_dir = Path('/content/drive/My Drive/Final Project/Fast food Dataset/Test')
test_filepath = list(test_dir.glob(r'**/*.jpeg'))
```

```python
valid_dir = Path('/content/drive/My Drive/Final Project/Fast food Dataset/Valid')
valid_filepath = list(valid_dir.glob(r'**/*.jpeg'))
```

```python
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
batch_size = 32
```

```python
# Generate training dataset
train_generator = datagen.flow_from_directory(
    train_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

# Generate validation dataset
validation_generator = datagen.flow_from_directory(
    valid_dir,
    target_size=(224, 224),
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

```

```python
# Define VGG16 model
vgg16_model = tf.keras.applications.VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
```

```python
# Freeze VGG16 layers
for layer in vgg16_model.layers:
    layer.trainable = False
```

```python
# Create a new model
model_vgg16 = tf.keras.Sequential([
    vgg16_model,
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(train_generator.num_classes, activation='softmax')
])
```

```python
# Compile the model
model_vgg16.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history_vgg16 = model_vgg16.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size,
    epochs=10
)
```

```python
history_df = pd.DataFrame(history_vgg16.history)
print(history_df[['accuracy', 'val_accuracy', 'loss', 'val_loss']] )


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history_vgg16.history['accuracy'], label='training accuracy')
plt.plot(history_vgg16.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(history_vgg16.history['loss'], label='training loss')
plt.plot(history_vgg16.history['val_loss'], label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

# **-- VGG 19 Model --**

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.applications import VGG19
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
from pathlib import Path
import os.path
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_data_dir = Path('/content/drive/My Drive/Final Project/Fast food Dataset/Train')
validation_data_dir = Path('/content/drive/My Drive/Final Project/Fast food Dataset/Valid')
epochs = 10
batch_size = 32
img_height, img_width = 224, 224
```

```python
# Use VGG19 as a base model
base_model = VGG19(weights='imagenet', include_top=False, input_shape=(img_height, img_width, 3))
```

```python
# Freeze the convolutional layers
for layer in base_model.layers:
    layer.trainable = False
```

```python
# Build the model
model = models.Sequential()
model.add(base_model)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)

validation_generator = test_datagen.flow_from_directory(
    validation_data_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
# Train the model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)
```

```python
import pandas as pd
from pathlib import Path
import os.path
import matplotlib.pyplot as plt

history_df = pd.DataFrame(history.history)
print(history_df[['accuracy', 'val_accuracy', 'loss', 'val_loss']] )


plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label='validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
